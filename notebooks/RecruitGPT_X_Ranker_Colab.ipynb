# RecruitGPT X — Offline Ranker (Colab Backup)

CPU-only, no-network ranker for the Redrob challenge. Upload `candidates.jsonl` or use the bundled sample.

```python
import sys, time
from pathlib import Path

REPO = Path('.')
if not (REPO / 'rank.py').exists():
    !git clone https://github.com/rahulkumarsinghj/recruitgpt-x.git
    %cd recruitgpt-x
    REPO = Path('.')

sys.path.insert(0, str(REPO))
from challenge.redrob_ranker import rank_candidates, write_submission

DATA = REPO / 'data' / 'candidates.jsonl'
if not DATA.exists():
    print('Upload candidates.jsonl to data/ or mount Drive')
else:
    t0 = time.perf_counter()
    top = rank_candidates(DATA, top_k=100)
    out = REPO / 'submission.csv'
    write_submission(top, out)
    print(f'Wrote {out} in {time.perf_counter()-t0:.1f}s')
    print('Top 3:')
    for i, r in enumerate(top[:3], 1):
        print(f'  {i}. {r.candidate_id} — {r.reasoning[:100]}...')
```
